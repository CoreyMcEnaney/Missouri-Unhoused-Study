**Intro**
I started this project because wanted to practice formatting datasets and writing a CREATE TABLE query in SQL. I uploaded and pulled data from thirty-two tables to gain some insight on the Missouri unhoused population and available beds, I created a new table with an sql query then a created a few visuals on Tableau.

**Data:**
Tables were downloaded from hudexchange.info https://www.hudexchange.info/resource/3031/pit-and-hic-data-since-2007/
* 2007 - 2023 PIT Estimates by State (XLSB)
* 2007 - 2023 HIC by State (XLSX)

Tables were converted to .csv then opened in SQL. A query was made resulting in the attached MO_Unhoused_Data.csv dataset
